```python
# Function to compute GCD using Euclid's Algorithm

#Problem 1

#Part 1
# Function to compute GCD using Euclid's Algorithm

def gcd(a, b):    #Assuming that the inputs are always positive(a,b > 0)
    while a != b: 
        if a > b:
            a = a - b
        else:
            b = b - a
    return a
print(gcd(500,1000))
#Part 2
print()
print("If one or both arguments are negative the function will display an error.")
```

```python
# Euler limit approximation

def euler_limit(n):
    return (1 + 1/n)**n
euler_limit(100)
```

```python
# floating point limitation occurs if n is a big value.
euler_limit(10**16)
```

```python
#PART 3
# Continue adding terms until they become smaller than epsilon

def factorial(n):
    final=1
    if n==0:
        return final
    for i in range(1,n+1):
        final*=i
    return final

def euler_sum(n):
    total=0
    for i in range(0,n+1,1):
        total += 1/factorial(i)
    return total

print(euler_sum(50))
```

```python

#Problem 2

#Part 4

def euler_approx(n:float):
    import math
    final=0
    m=0
    while math.exp(1)-euler_sum(m)> n:
        m= m+1
        final= euler_sum(m)
    return final

print(euler_approx(0.5))
```

```python
# Check if n is a power of 2
def is_power_of_2(n):
    if n==1:
        return True
    while n>1:
       if n%2!=0:
            return False 
       n = n//2
    return n==1
is_power_of_2(8)
```

```python
# Check if n is a power of base b
def is_power(b,n):
    if b==1:
        return n==1
    if n==1:
        return True
    while n>1:
        if n%b!=0:
            return False
        n = n//b
    return n==1
is_power(5,125)
```
